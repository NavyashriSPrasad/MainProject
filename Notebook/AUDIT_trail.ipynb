```python
import pandas as pd

df=pd.read_csv('AUDIT_dataset - 1.csv')
df.head()
```

```python
df.drop('Participant ID', axis=1, inplace=True)
df.head()
```

```python
df.rename(columns={'How often do you have a drink containing alcohol?': 'Q1', 'How many standard drinks do you have on a typical drinking day?': 'Q2' , 'How often do you have six or more drinks on one occasion?':'Q3'
                  ,'How often do you find it hard to stop drinking once started?':'Q4','How often do you fail to do what is expected due to drinking?':'Q5',
                  'How often do you need a drink in the morning after heavy drinking?':'Q6','How often do you feel guilt or remorse after drinking?':'Q7',
                  'How often do you forget what happened because of drinking?':'Q8','Have you or someone else been injured due to your drinking?':'Q9','Has someone been concerned about your drinking or suggested cutting down?':'Q10'}, inplace=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Q1']=le.fit_transform(df['Q1'])
df['Q2']=le.fit_transform(df['Q2'])
df['Q3']=le.fit_transform(df['Q3'])
df['Q4']=le.fit_transform(df['Q4'])
df['Q5']=le.fit_transform(df['Q5'])
df['Q6']=le.fit_transform(df['Q6'])
df['Q7']=le.fit_transform(df['Q7'])
df['Q8']=le.fit_transform(df['Q8'])
df['Q9']=le.fit_transform(df['Q9'])
df['Q10']=le.fit_transform(df['Q10'])
```

```python
df.head()
```

```python
df['Level']=le.fit_transform(df['Level'])
df.head()
```

```python
db=df.copy()
db.head()
```

```python
db.drop('Total Score', axis=1, inplace=True)
```

```python
db.head()
```

```python
from sklearn.model_selection import train_test_split
x=db[['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9','Q10']]
y=db['Level']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,shuffle=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# SVM model training
svm_model = SVC(kernel='linear')  # You can try different kernels ('linear', 'rbf', 'poly', etc.)
svm_model.fit(xtrain, ytrain)


predictions = svm_model.predict(xtest)
accuracy = accuracy_score(ytest, predictions)
report = classification_report(ytest, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
svm = SVC(C=100, gamma=0.1)

svm.fit(xtrain, ytrain)

# Make predictions on the test/validation set
predictions = svm.predict(xtest)

# Evaluate the model's performance
accuracy = accuracy_score(ytest, predictions)
print(f"Accuracy on Test/Validation Set: {accuracy:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import make_scorer, accuracy_score

# Define the SVM model
svm = SVC()

# Define the hyperparameters and their values for tuning
param_grid = {
    'C': [0.1, 1, 10, 100],  
    'gamma': [0.1, 1, 10]    
}

# scoring metric
scorer = make_scorer(accuracy_score)

# GridSearchCV
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, scoring=scorer, cv=3)
grid_result = grid_search.fit(xtrain, ytrain)

# Get the best parameters and best score
best_params = grid_result.best_params_
best_score = grid_result.best_score_

print("Best Parameters:", best_params)
print("Best Accuracy:", best_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)  # You can set different hyperparameters here
dt_model.fit(xtrain, ytrain)

# Model evaluation
predictions = dt_model.predict(xtest)
accuracy = accuracy_score(ytest, predictions)
report = classification_report(ytest, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
from sklearn.model_selection import GridSearchCV


param_grid = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}


dt = DecisionTreeClassifier(random_state=42)


grid_search = GridSearchCV(estimator=dt, param_grid=param_grid, cv=5, scoring='accuracy')
grid_search.fit(xtrain, ytrain)

best_params = grid_search.best_params_
best_model = grid_search.best_estimator_

predictions = best_model.predict(xtest)
accuracy = accuracy_score(ytest, predictions)
report = classification_report(ytest, predictions)

print(f"Best Parameters: {best_params}")
print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
# Mentioning class weights
class_weights = {0:1,1: 1,2: 5} 

#  Decision Tree classifier with class weights
dt = DecisionTreeClassifier(random_state=42, class_weight=class_weights)


dt.fit(xtrain, ytrain)
predictions = dt.predict(xtest)
accuracy = accuracy_score(ytest, predictions)
report = classification_report(ytest, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")

```

```python
from sklearn.ensemble import RandomForestClassifier


rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  

rf_model.fit(xtrain, ytrain)

rf_predictions = rf_model.predict(xtest)

rf_accuracy = accuracy_score(ytest, rf_predictions)
rf_report = classification_report(ytest, rf_predictions)

print(f"Random Forest Accuracy: {rf_accuracy}")
print(f"Random Forest Classification Report:\n{rf_report}")
```
