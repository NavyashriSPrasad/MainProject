```python
import pandas as pd

df=pd.read_csv('AUDIT_dataset - 1.csv')
df.head(10)
```

```python
df.drop('Participant ID', axis=1, inplace=True)
df.head(10)
```

```python
df.rename(columns={'How often do you have a drink containing alcohol?': 'Q1', 'How many standard drinks do you have on a typical drinking day?': 'Q2' , 'How often do you have six or more drinks on one occasion?':'Q3'
                  ,'How often do you find it hard to stop drinking once started?':'Q4','How often do you fail to do what is expected due to drinking?':'Q5',
                  'How often do you need a drink in the morning after heavy drinking?':'Q6','How often do you feel guilt or remorse after drinking?':'Q7',
                  'How often do you forget what happened because of drinking?':'Q8','Have you or someone else been injured due to your drinking?':'Q9','Has someone been concerned about your drinking or suggested cutting down?':'Q10'}, inplace=True)
```

```python
df.head()
```

```python
import pandas as pd




# Define custom encoding mappings for each question
custom_mappings = {
    'Q1': {'Never': 0, 'Monthly or less': 1, '2-4 times a month': 2, '2-3 times a week': 3, '4 or more times a week': 4},
    'Q2': {'1 or 2': 0, '3 or 4': 1, '5 or 6': 2, '7 to 9': 3, '10 or more': 4},
    'Q3': {'Never': 0, 'Less than monthly': 1, 'Monthly': 2, 'Weekly': 3, 'Daily or almost daily': 4},
    'Q4': {'Never': 0, 'Less than monthly': 1, 'Monthly': 2, 'Weekly': 3, 'Daily or almost daily': 4},
    'Q5': {'Never': 0, 'Less than monthly': 1, 'Monthly': 2, 'Weekly': 3, 'Daily or almost daily': 4},
    'Q6': {'Never': 0, 'Less than monthly': 1, 'Monthly': 2, 'Weekly': 3, 'Daily or almost daily': 4},
    'Q7': {'Never': 0, 'Less than monthly': 1, 'Monthly': 2, 'Weekly': 3, 'Daily or almost daily': 4},
    'Q8': {'Never': 0, 'Less than monthly': 1, 'Monthly': 2, 'Weekly': 3, 'Daily or almost daily': 4},
    'Q9': {'No': 0, 'Yes, but not in the past year': 2,'Yes, during the past year':4},
    'Q10': {'No': 0, 'Yes, but not in the past year': 2,'Yes, during the past year':4}
}

for col in df.columns:
    if col in custom_mappings:
        df[col] = df[col].map(custom_mappings[col])

# Display the encoded DataFrame
print(df)
```

```python
data = {
    'Q1': ['never',  'monthly', '2-4 times a month', '2-3 times a week','4 or more times a week'],
    'Q2': ['1 or 2', '3 or 4', '5 or 6', '7 to 9', '10 or more'],
    'Q3': ['never', 'less than monthly', 'monthly', 'weekly', 'daily or almost daily'],
    'Q4': ['never', 'less than monthly', 'monthly', 'weekly', 'daily or almost daily'],
    'Q5': ['never', 'less than monthly', 'monthly', 'weekly', 'daily or almost daily'],
    'Q6': ['never', 'less than monthly', 'monthly', 'weekly', 'daily or almost daily'],
    'Q7': ['never', 'less than monthly', 'monthly', 'weekly', 'daily or almost daily'],
    'Q8': ['never', 'less than monthly', 'monthly', 'weekly', 'daily or almost daily'],
    'Q9': ['no', 'Yes, but not in the past year', 'Yes, during the past year'],
    'Q10': ['no', 'Yes, but not in the past year', 'Yes, during the past year']
}
```

```python
df.drop('Total Score', axis=1, inplace=True)
```

```python
df.head()
```

```python

# Calculate total score (sum of each row)
df['Total_Score'] = df.sum(axis=1)

# Display the DataFrame with the new column
print(df)
```

```python
df.drop('Level', axis=1, inplace=True)
```

```python




low_risk_cutoff = 7
hazardous_harmful_cutoff = 14

# Assign risk levels based on total score
df['Risk Level'] = pd.cut(df['Total_Score'], bins=[-float('inf'), low_risk_cutoff, hazardous_harmful_cutoff, float('inf')],
                          labels=['Low Risk', 'Hazardous', 'Alcohol Dependence'])

# Display the DataFrame with the new column
df.head(30)
```

```python
from sklearn.model_selection import train_test_split
x= df[['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']]
y=df['Risk Level']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# SVM model training
svm_model = SVC(kernel='linear')  # You can try different kernels ('linear', 'rbf', 'poly', etc.)
svm_model.fit(X_train, y_train)


predictions = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
svm = SVC(C=100, gamma=0.1)

svm.fit(X_train, y_train)

# Make predictions on the test/validation set
predictions = svm.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy on Test/Validation Set: {accuracy:.4f}")
```

```python
new_data=[[0,0,0,0,1,0,1,0,1,1]]
predicted_class = svm.predict(new_data)

print("Predicted Class:", predicted_class)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import make_scorer, accuracy_score

# Define the SVM model
svm = SVC()

# Define the hyperparameters and their values for tuning
param_grid = {
    'C': [0.1, 1, 10, 100],  
    'gamma': [0.1, 1, 10]    
}

# scoring metric
scorer = make_scorer(accuracy_score)

# GridSearchCV
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, scoring=scorer, cv=3)
grid_result = grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_result.best_params_
best_score = grid_result.best_score_

print("Best Parameters:", best_params)
print("Best Accuracy:", best_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)  # You can set different hyperparameters here
dt_model.fit(X_train, y_train)

# Model evaluation
predictions = dt_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust hyperparameters here

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
rf_predictions = rf_model.predict(X_test)

# Evaluate the model
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_report = classification_report(y_test, rf_predictions)

print(f"Random Forest Accuracy: {rf_accuracy}")
print(f"Random Forest Classification Report:\n{rf_report}")
```

```python
new_data=[[3,0,1,0,1,2,4,0,1,4]]
predicted_class = rf_model.predict(new_data)

print("Predicted Class:", predicted_class)
```
