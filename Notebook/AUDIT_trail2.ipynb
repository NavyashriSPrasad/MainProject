```python
import pandas as pd

df=pd.read_csv('AUDIT_dataset - 1.csv')
df.head(10)
```

```python
df.drop('Participant ID', axis=1, inplace=True)
df.head(10)
```

```python
df.rename(columns={'How often do you have a drink containing alcohol?': 'Q1', 'How many standard drinks do you have on a typical drinking day?': 'Q2' , 'How often do you have six or more drinks on one occasion?':'Q3'
                  ,'How often do you find it hard to stop drinking once started?':'Q4','How often do you fail to do what is expected due to drinking?':'Q5',
                  'How often do you need a drink in the morning after heavy drinking?':'Q6','How often do you feel guilt or remorse after drinking?':'Q7',
                  'How often do you forget what happened because of drinking?':'Q8','Have you or someone else been injured due to your drinking?':'Q9','Has someone been concerned about your drinking or suggested cutting down?':'Q10'}, inplace=True)
```

```python
df.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Q1']=le.fit_transform(df['Q1'])
df['Q2']=le.fit_transform(df['Q2'])
df['Q3']=le.fit_transform(df['Q3'])
df['Q4']=le.fit_transform(df['Q4'])
df['Q5']=le.fit_transform(df['Q5'])
df['Q6']=le.fit_transform(df['Q6'])
df['Q7']=le.fit_transform(df['Q7'])
df['Q8']=le.fit_transform(df['Q8'])
df['Q9']=le.fit_transform(df['Q9'])
df['Q10']=le.fit_transform(df['Q10'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'numeric_columns' contains the names of numerical columns in your dataset
numeric_columns = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10']

# Original data
original_data = df[numeric_columns]

# Transformed data (logarithmic transformation)
transformed_data = np.log1p(original_data)

# Set up subplots
fig, axes = plt.subplots(nrows=2, ncols=len(numeric_columns), figsize=(15, 6))

# Plot histograms for original data
for i, col in enumerate(numeric_columns):
    sns.histplot(original_data[col], kde=True, ax=axes[0, i])
    axes[0, i].set_title(f'Original {col}')

# Plot histograms for transformed data
for i, col in enumerate(numeric_columns):
    sns.histplot(transformed_data[col], kde=True, ax=axes[1, i])
    axes[1, i].set_title(f'Transformed {col}')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
y=df['Level']
X_train, X_test, y_train, y_test = train_test_split(transformed_data, y, test_size=0.2, random_state=42)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# SVM model training
svm_model = SVC(kernel='linear')  # You can try different kernels ('linear', 'rbf', 'poly', etc.)
svm_model.fit(X_train, y_train)


predictions = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
svm = SVC(C=100, gamma=0.1)

svm.fit(X_train, y_train)

# Make predictions on the test/validation set
predictions = svm.predict(X_test)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy on Test/Validation Set: {accuracy:.4f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import make_scorer, accuracy_score

# Define the SVM model
svm = SVC()

# Define the hyperparameters and their values for tuning
param_grid = {
    'C': [0.1, 1, 10, 100],  
    'gamma': [0.1, 1, 10]    
}

# scoring metric
scorer = make_scorer(accuracy_score)

# GridSearchCV
grid_search = GridSearchCV(estimator=svm, param_grid=param_grid, scoring=scorer, cv=3)
grid_result = grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_result.best_params_
best_score = grid_result.best_score_

print("Best Parameters:", best_params)
print("Best Accuracy:", best_score)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)  # You can set different hyperparameters here
dt_model.fit(X_train, y_train)

# Model evaluation
predictions = dt_model.predict(X_test)
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)

print(f"Accuracy: {accuracy}")
print(f"Classification Report:\n{report}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Instantiate the Random Forest classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)  # You can adjust hyperparameters here

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions
rf_predictions = rf_model.predict(X_test)

# Evaluate the model
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_report = classification_report(y_test, rf_predictions)

print(f"Random Forest Accuracy: {rf_accuracy}")
print(f"Random Forest Classification Report:\n{rf_report}")
```

```python
new_data=[[3,0,1,0,1,2,4,0,1,4]]
predicted_class = rf_model.predict(new_data)

print("Predicted Class:", predicted_class)
```

```python
new_data=[[0,0,0,0,1,2,0,0,1,0]]
predicted_class = rf_model.predict(new_data)

print("Predicted Class:", predicted_class)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

# Initialize Random Forest as the base estimator
base_estimator = RandomForestClassifier(n_estimators=100, random_state=42)

# Initialize AdaBoost classifier with Random Forest as base estimator
adaboost_classifier = AdaBoostClassifier(base_estimator=base_estimator, n_estimators=50, random_state=42)

# Train the AdaBoost classifier
adaboost_classifier.fit(X_train, y_train)

# Make predictions
adaboost_predictions = adaboost_classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, adaboost_predictions)
print(f"AdaBoost with Random Forest Accuracy: {accuracy}")

# Generate classification report
report = classification_report(y_test, adaboost_predictions)
print(f"Classification Report:\n{report}")
```
